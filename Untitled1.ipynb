```python
import tensorflow as tf
from tensorflow import keras
```

#Data definition


```python
x = tf.constant([-1, -2, 0, 1, 2, 3, 7], dtype=float)
y = x * 3 -5
```

```python
x.numpy()
y.numpy()
```

#Model definition


```python
model = keras.Sequential([
    keras.layers.Dense(units=1, activation=None, input_shape =[1])
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
model.summary()
```

```python
model.fit(x, y, batch_size=1, epochs=500)
```

#Test

```python
x_test = tf.constant([-4, 11, 20], dtype=float)
y_test = x_test * 3 - 5
```

```python
x_test.numpy()
y_test.numpy()
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_test.numpy(), y_test.numpy(), color='red', linewidths=8)
plt.scatter(x_test.numpy(), y_pred, color='yellow', linewidths=2)
plt.show
```
