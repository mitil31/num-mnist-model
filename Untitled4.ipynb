```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-80, 81, 0.5)
y = x**2 + 5 * x +3
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape, x_test.shape
```

```python
model_2 = keras.Sequential()
model_2.add(keras.layers.Dense(units=256, activation='relu', input_shape = [1]))
model_2.add(keras.layers.Dense(units=256, activation='relu'))
model_2.add(keras.layers.Dense(units=256, activation='relu'))
model_2.add(keras.layers.Dense(units=256, activation='relu'))
model_2.add(keras.layers.Dense(units= 1))
```

```python
optimizer = keras.optimizers.RMSprop(learning_rate=0.0001)
loss = keras.losses.MeanSquaredError
metric = ["root_mean_squared_error"]
model_2.compile(optimizer= optimizer, loss= loss, metrics= metric)
```

```python
early_stop = keras.callbacks.EarlyStopping(monitor= 'val_loss', patience= 10, restore_best_weights= True)
```

```python
hist_2 = model_2.fit(x_train, y_train, epochs= 1000, batch_size= 128, callbacks= [early_stop], validation_data= (x_test, y_test))
```

```python
plt.plot(hist_2.history['loss'], color= 'red')
plt.plot(hist_2.history['val_loss'], color= 'blue')
plt.show()
```
